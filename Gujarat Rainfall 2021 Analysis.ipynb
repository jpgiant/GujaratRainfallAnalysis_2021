```python
region=['Kutch','Saurashtra','North Gujarat','Central Gujarat','South Gujarat']
```

```python
def classifier(row):
    if row['Deviation from 100%'] >=60:
        return 'Large Excess'
    elif (row['Deviation from 100%'] < 60) & (row['Deviation from 100%'] >= 20):
        return 'Excess'
    elif (row['Deviation from 100%'] < 20) & (row['Deviation from 100%'] >= -20):
        return 'Normal'
    elif (row['Deviation from 100%'] < -20) & (row['Deviation from 100%'] >= -60):
        return 'Deficient'
    elif (row['Deviation from 100%'] < -60) & (row['Deviation from 100%'] >= -100):
        return 'Large Deficient'
    else:
        return 'No Data'
```

```python
data['Classification']=data.apply(classifier,axis=1)
```

```python
#Data classified as per guidelines of Indian Meteorological Department
data
```

```python
data
```

```python
# Unique values in Classification column
data['Classification'].unique()
```

```python
norm=data[data['Classification']=='Normal']
defic=data[data['Classification']=='Deficient']
excess=data[data['Classification']=='Excess']
print('Districts having Excess Rainfall:',len(excess))
print('Districts having Normal Rainfall:',len(norm))
print('Districts having Deficient Rainfall:',len(defic))
```

```python
# Percentage of Districts as per their Classification
excess_pc=np.round((len(excess)/len(data))*100,decimals=2)
norm_pc=np.round((len(norm)/len(data))*100,decimals=2)
defic_pc=np.round((len(defic)/len(data))*100,decimals=2)
print('Percentage districts with Excess Rainfall:',excess_pc,'%')
print('Percentage districts with Normal Rainfall:',norm_pc,'%')
print('Percentage districts with Deficit Rainfall:',defic_pc,'%')
```
